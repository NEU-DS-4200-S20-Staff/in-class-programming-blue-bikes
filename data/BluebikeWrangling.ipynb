```python
# Packages
import pandas as pd

station_data = pd.read_csv('Hubway_Stations_as_of_July_2017.csv')
ride_data = pd.read_csv('201903-bluebikes-tripdata.csv')
```

```python
station_data.head()
```

```python
ride_data.head()
```

```python
start_counts = pd.DataFrame(ride_data['start station name'].value_counts())
start_counts.reset_index(level=0, inplace=True)
start_counts.columns=('start station name', 'trip start count')
end_counts = pd.DataFrame(ride_data['end station name'].value_counts())
end_counts.reset_index(level=0, inplace=True)
end_counts.columns=('end station name', 'trip end count')

start_counts.head()
end_counts.head()
```

```python
station_counts = station_data.merge(start_counts, left_on='Name', right_on='start station name')
station_counts = station_counts.merge(end_counts, left_on='Name', right_on='end station name')
station_counts.head()
```

```python
station_counts
```

```python
station_counts.to_csv("march2017_bluebikes_trip_counts.csv")
```
